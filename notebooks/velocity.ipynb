```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('data/velocities_10000K.txt')
vx, vy = data[:,1], data[:,2]               

kB = 1.380649e-23                           
m  = 12.0107 * 1.6605390666e-27              
T  = 10000                                  

vx *= 1e2
vy *= 1e2
v  = np.sqrt(vx**2 + vy**2)

vmax = 1.1*np.max(np.abs(np.concatenate((vx, vy))))
sigma  = np.sqrt(kB*T/m)                     # m/s
xs     = np.linspace(-vmax, vmax, 400)

plt.figure(figsize=(6,4))
plt.hist(vx, bins=100, density=True, alpha=0.5, label='vx  (data)')
plt.hist(vy, bins=100, density=True, alpha=0.5, label='vy  (data)')
plt.plot(xs, 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-xs**2/(2*sigma**2)),
         'k-', lw=2, label='Gaussian  $\\sigma=\\sqrt{k_BT/m}$')
plt.xlabel('$v_x$  or  $v_y$  (m s$^{-1}$)')
plt.ylabel('probability density')
plt.legend()
plt.tight_layout()
plt.show()

vs = np.linspace(0, vmax, 400)
plt.figure(figsize=(6,4))
plt.hist(v, bins=100, density=True, alpha=0.6, label='speed  (data)')
plt.plot(vs, (m/(kB*T))*vs*np.exp(-m*vs**2/(2*kB*T)),
         'k-', lw=2, label='Rayleigh  (2D MB)')
plt.xlabel('$v = \\sqrt{v_x^2+v_y^2}$  (m s$^{-1}$)')
plt.ylabel('probability density')
plt.legend()
plt.tight_layout()
plt.show()
```
