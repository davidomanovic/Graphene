# Single trajectory

```python
a = 1.42 # lattice parameter for graphene

pos = np.loadtxt(f'data/hexatic.data', skiprows=15)[:, 2:4] / a # unitless
noise = np.random.normal(loc=0.0, scale=0.05, size=pos.shape)
pos += noise # adding tiny perturbation for phonon contributions

ax = plot_g6(pos, dr = 0.05, rmin=1.0, rmax=60, colors='green')
```

# Multiple trajectories
